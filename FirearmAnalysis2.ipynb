## About Dataset

The data was sourced from https://usafacts.org and focuses on firearm-related fatalities in the United States spanning from 1981 to 2021. This dataset holds significance as it allows for an examination of the necessity for gun-related laws and the common belief that having firearms at home provides a sense of security.

Exploring this dataset presents an opportunity to delve into the factors contributing to firearm incidents and the  causes behind them.This will helps us in fostering good relationship with the soceity. 

#### Data Description

Most variables are named in quite obvious ways.

Description of the variables:
- Years: The data is for which year
- Firarm Deaths: Total number of firearms deaths in that year
- Homicide: Number of deaths because of Homicide
- Suicide: Number of deaths because of Suicide
- Unintentional: Number of deaths because of Unintentional
- Legal Inetervention/Operations of War: Deaths because of Legal Internventions
- Undetermined: Deaths for which we don't know the reason for the death
- Under 19: Number of deaths where the victim was under19
- 20-34: Age is under 20-34
- 35-49: Age is under 35-49
- 50-64: Age is under 50-64
- 65+: Age is over 65+
- Male: Number of firearm death for the male gender
- Female: Number of firearm deaths for the female gender
- State Wise Count: Firearm deaths for each state across each year

### Import Libraries

```python
import pandas as pd
import altair as alt
```

### Read Dataset

```python
firearm= pd.read_csv('firearm_deaths_usafacts (1).csv')
```

```python
# Store the year in pandas year format
firearm['Years']=pd.to_datetime(firearm['Years'],format='%Y').dt.year
```

In this analysis, we are examining the various factors contributing to firearm-related deaths, focusing our research on the specific contexts of New York and Texas. These states have higher crime rates and motivates to conduct an in-depth investigation into the fundamental causes of this concerning trend.

As we navigate through the data, our objective is to identify the peaks in death counts; but also to understand the average firearm-related fatalities in both Texas and New York. These statistics is important in comprehending the gravity of the issue, provide us with a broader perspective on the severity of the situation.



```python
firearm['Texas_Contribution']= (firearm['Texas']/firearm['Firearmdeaths'])*100
firearm['NewYork_Contribution']= (firearm['NewYork']/firearm['Firearmdeaths'])*100
```

```python
## Cause of Firearm Deaths Over Time

cause_firearm_time = alt.Chart(firearm).transform_fold(
    ['NewYork','Texas'],
    as_=['State','Count']
).mark_line().encode(
    alt.X('Years:O',
           axis = alt.Axis(labelAngle=0,title='Year',labelFontSize = 12,
                         titleFontSize = 12,
                         values=[1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020] ),
           #bin=alt.BinParams(step=5)
         ),
    alt.Y('Count:Q',
           axis = alt.Axis(title='Number Of Firearm Deaths',labelFontSize = 15,
                         titleFontSize = 12)),
    alt.Color('State:N',legend=None),
    text=alt.Text('State:N'),
    tooltip =['State:N','Years:O','Count:Q']
    
).properties(
    title = alt.Title(text='Firearm Death Count By State',fontSize=20)
).properties(
    width=800,
    height=300,
    title = alt.Title(text='Firearm Trend from (1981-2021)',fontSize=15)
)


labels_NY = alt.Chart(firearm).transform_fold(
   ['NewYork'],
    as_=['State','Count']
).mark_text(align='left', dx=6,dy = -15,size = 13).encode(
    alt.X('Years:O', aggregate='max'),
    alt.Y('Count:Q', aggregate='min'),
    alt.Text('State:N'),
    
    alt.Color('State:N', legend=None)
)

labels_TX = alt.Chart(firearm).transform_fold(
   ['Texas'],
    as_=['State','Count']
).mark_text(align='left', dx=6,dy = -15,size = 13).encode(
    alt.X('Years:O', aggregate='max'),
    alt.Y('Count:Q', aggregate='max'),
    alt.Text('State:N'),
    alt.ColorValue('#ff7f0e')
)

## Now let's also show the minimum and maximum count on the plot

mean_line_TX = alt.Chart(firearm).mark_rule(
        stroke='#ff7f0e',strokeDash=[2,4],
        size = 3
        ).encode(
            alt.Y('Texas:Q',
                  aggregate='average'))


mean_line_NY = alt.Chart(firearm).mark_rule(
        stroke='#1f77b4',strokeDash=[2,4],
        size = 3
    ).encode(alt.Y('NewYork:Q',
             aggregate='average'))




# Adding text labels for each point (representing the years)
labels_mean_tx= mean_line_TX.mark_text(
    align='left',
    dx=250,  
    dy=-10, 
    fontSize=10
).encode(
    text=alt.value('Texas mean firearm')
)

# Adding text labels for each point (representing the years)
labels_mean_ny= mean_line_TX.mark_text(
    align='left',
    dx=240,  
    dy=90,  
    fontSize=10
).encode(
    text=alt.value('New York mean firearm')  
)



line_chart_state = alt.layer(cause_firearm_time,labels_NY,labels_TX,mean_line_TX,mean_line_NY,labels_mean_tx,labels_mean_ny).resolve_scale(color='independent')
line_chart_state




total_state = alt.Chart(firearm).mark_bar().transform_fold(
   ['NewYork','Texas'],
    as_=['State','Count']
).transform_aggregate(
    sum_count='sum(Count)',
    groupby=['State']
).encode(
    alt.X('sum_count:Q',
          axis =alt.Axis(title='Number of Firearm Deaths',
                         labelFontSize = 12,
                         titleFontSize = 12)),
    alt.Y('State:N', 
          axis = alt.Axis(labelFontSize=12,
                          title='State',titleFontSize=12),
          sort = ['NewYork','Texas'],),  
 
    tooltip=['sum_count:Q', 'State:N'],
    
).properties(
    width=700,
    height=300,
    title = alt.Title(text='Firearm Death Total',fontSize=15)
)


# Text labels for the sum of counts for Texas and New York
labels_count_TX = alt.Chart(firearm).mark_text(
    align='left', dx=6, dy=-15, size=13, fontWeight='bold'
).encode(
    x='sum(Texas):Q',
    y=alt.value(245),  
    text=alt.Text('sum(Texas):Q', format=','),
    color=alt.value('black')  
)

labels_count_NY= alt.Chart(firearm).mark_text(
    align='left', dx=6, dy=-15, size=13, fontWeight='bold'
).encode(
    x='sum(NewYork):Q',
    y=alt.value(80),  
    text=alt.Text('sum(NewYork):Q', format=','),
    color=alt.value('black')  
)

# Layer the bar chart and the text labels
total_state_with_labels = alt.layer(
    total_state, 
    labels_count_TX,
    labels_count_NY
).properties(width = 350)


mean_firearm_death_tx = firearm['Texas'].mean()
mean_firearm_death_ny = firearm['NewYork'].mean()


cor_texas_ny = alt.Chart(firearm).mark_point(
    filled=True,
    color ='grey'
).encode(
    alt.X('Texas:Q', title='Firearm Deaths in Texas'),
    alt.Y('NewYork:Q', title='Firearm Deaths in New York'),
    tooltip=['Texas:Q', 'NewYork:Q', 'Years:O']
)



# Create rectangles to highlight areas where firearm deaths are more than the mean
highlighted_areas = alt.Chart(firearm).mark_rect(
    color='green',
    opacity=0.1
).encode(
    x='mean(Texas):Q',
    x2='max(Texas):Q',
    y='mean(NewYork):Q',
    y2='max(NewYork):Q',
    
)



# Layer the scatter plot with the highlighted areas
layered_scatter_plot = (cor_texas_ny   + highlighted_areas).properties(
    title='Firearm Deaths Comparison: Texas vs. New York'
)




mean_NY = alt.Chart(firearm).mark_rule(
        stroke='black',#strokeDash=[1,2],
        size = 1
    ).encode(alt.Y('NewYork:Q',
             aggregate='average'))


mean_TX = alt.Chart(firearm).mark_rule(
        stroke='black',#strokeDash=[1,2],
        size = 1
    ).encode(alt.X('Texas:Q',
             aggregate='average'))
#layered_scatter_plot+mean_NY+mean_TX

# Layer the scatter plot with the highlighted areas
layered_scatter_plot = (cor_texas_ny   + highlighted_areas+mean_NY+mean_TX).properties(
    
    title = alt.Title(text='Firearm Deaths Comparison',fontSize=15)
)


layered_scatter_plot

legend_chart = alt.Chart(firearm).mark_text(align='left',opacity=0.2).encode(
    color=alt.Color('Area:N', scale=alt.Scale(domain=['Firearm > Mean'], range=['green']))
)
layered_scatter_plot = legend_chart+layered_scatter_plot



contribution_charts =alt.Chart(firearm).mark_line(
).encode(
    alt.X('Years:O',
           axis = alt.Axis(labelAngle=0,title='Year',labelFontSize = 12,
                         titleFontSize = 12,
                         values=[1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020] ),),
    alt.Y(alt.repeat('column'), type='quantitative',
          #scale=alt.Scale(domain=['Texas_Contribution', 'NewYork_Contribution'], range=['blue', 'red'])
         )
).properties(
  width=372,
  height=180,
  title='Firearm Death Contribution Over Years'
    
).repeat(
  column=['Texas_Contribution', 'NewYork_Contribution']
).resolve_scale(y='shared')




ny_texas = alt.vconcat(alt.hconcat(total_state_with_labels,layered_scatter_plot),contribution_charts).configure_view(
    stroke='transparent'
).configure_legend(
    labelFontSize=15,  
    titleFontSize=15  ,
    labelOffset=10,
    legendX=500,
    offset = 50,
    orient='right'
    
).resolve_scale('independent').properties(
    config={'legend': {'labelFontSize': 10, 
                       'titleFontSize': 15 ,
                       'padding': 0,
                       'labelOffset':1,
                       'offset':0.1
                      }},
   title = {
        "text" : ['Comparative Analysis Of Firearm Deaths: Texas VS New York'],
        #"subtitle":['1981 and 2021 Year'],
        "dx":180,
        "fontSize":20,
        "offset":40,
        "subtitlePadding":10,
        "subtitleFontSize":18
    
    }
)
ny_texas.save('ny_texas.html')
```

### Observations:  

- A noticeable disparity in firearm-related deaths is evident between Texas and New York. Texas has a significantly higher firearm death toll, totaling 123,104, compared to New York's 56,107.
- It's noteworthy that the number of firearm-related deaths is increasing in both Texas and New York. The trends in these two states appear to be correlated. However, there is a significant concern for Texas, as the firearm death rate shows a substantial increase with each passing year.
- To provide a clearer visualization of the trend, up until 2014, the patterns in Texas and New York were quite similar. Both states experienced a peak in the early 90s, followed by a decline.
- New York has effectively kept its firearm-related deaths below the mean firearm death rate, even now. In contrast, Texas surpassed this threshold after 2015, experiencing a concerning and rapid rise in firearm-related deaths.
- This stark difference may be attributed to the varying nature of firearm regulations in these two states. New York maintains stringent gun laws, while Texas is more lenient, making it relatively easy to acquire firearms, sometimes even from retail stores like Walmart.
- In the scatter plot, the highlighted green area represents a period when firearm deaths in both New York and Texas exceeded the average firearm deaths for both states. During this period, both states experienced a significant spike in firearm-related fatalities.
- In the final chart, we observe the contributions of New York and Texas to firearm-related deaths over time. Notably, Texas contributes more significantly to these deaths.
- It's worth noting that New York's contribution has decreased over the years, which is a positive trend.
